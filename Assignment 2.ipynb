```python
import random

# Player 1
print("--- Player 1's Turn ---")
dealer_num1 = random.randint(0, 10) #generate a randon number between 0 and 10
tries1 = 0  #This variable store the tries , set default to 0

while True:  #Loop will run untill correct num is entered
    guess = int(input("Player1: ")) 
    tries1 += 1 #increment in number of tries.
    
    if guess < dealer_num1:
        print("Try a greater number.")
    elif guess > dealer_num1:
        print("Try a smaller number.")
    else:
        print("That's right! Number of tries :", tries1)
        break  #exit the loop if the guess is correct.

# Player 2 
print("\n--- Player 2's Turn ---")
dealer_num2 = random.randint(0, 10)
tries2 = 0

while True:
    guess = int(input("Player2: "))
    tries2 += 1
    
    if guess < dealer_num2:
        print("Try a greater number.")
    elif guess > dealer_num2:
        print("Try a smaller number.")
    else:
        print("That's right! Number of tries :", tries2)
        break  

# Determine the winner
print("\n")
if tries1 < tries2:
    print(f"Winner is Player 1")
elif tries2 < tries1:
    print(f"Winner is Player 2")
else:
    print(f"It's a tie! (Both took {tries1} tries)")
```
