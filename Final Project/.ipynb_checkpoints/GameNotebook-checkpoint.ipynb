## Introduction

This project implements a complete Tic-Tac-Toe game in Python.

It includes:
- A console-based version (text mode)
- A graphical version using Tkinter (UI)
- Winner and draw detection
- Input validation
- A reset option in the UI

The project follows the assignment requirements: it uses functions, includes testing, and provides a user interface for bonus points.

```python
# Tic‑Tac‑Toe Game (Console Version)

# The board is a list of 9 positions (0–8)
board = [" " for _ in range(9)]

# Function to display the board in a 3×3 layout
def display_board():
    print("\n")
    print(board[0], "|", board[1], "|", board[2])
    print("--+---+--")
    print(board[3], "|", board[4], "|", board[5])
    print("--+---+--")
    print(board[6], "|", board[7], "|", board[8])
    print("\n")

# All winning combinations (rows, columns, diagonals)
winning_positions = [
    (0,1,2), (3,4,5), (6,7,8),   # rows
    (0,3,6), (1,4,7), (2,5,8),   # columns
    (0,4,8), (2,4,6)             # diagonals
]

# Function to check if a player has won
def check_winner(player):
    # Check each winning combination
    for a, b, c in winning_positions:
        if board[a] == board[b] == board[c] == player:
            return True
    return False

# Function to check if the board is full (draw)
def check_draw():
    return " " not in board

# Main game loop
current_player = "X"   # X always starts

while True:
    display_board()

    # Ask player for a position (1–9)
    move = input(f"Player {current_player}, choose a position (1-9): ")

    # Validate input (must be digit and within range)
    if not move.isdigit() or not (1 <= int(move) <= 9):
        print("Invalid input. Try again.")
        continue

    move = int(move) - 1  # Convert to 0–8 index

    # Check if the chosen spot is empty
    if board[move] != " ":
        print("That spot is taken. Try again.")
        continue

    # Place the player's mark
    board[move] = current_player

    # Check for win
    if check_winner(current_player):
        display_board()
        print(f"Player {current_player} wins!")
        break

    # Check for draw
    if check_draw():
        display_board()
        print("It's a draw!")
        break

    # Switch player
    current_player = "O" if current_player == "X" else "X"
```

## Tic-Tac-Toe UI Version (Tkinter)

To earn the bonus points, this project also includes a graphical user interface (UI) using Tkinter.

The UI:
- Shows a 3×3 grid of buttons
- Lets players click to place X or O
- Detects wins and draws
- Provides a Reset button to start a new game

```python
import tkinter as tk
from tkinter import messagebox

# --- Game Logic ---

def reset_board():
    # Reset board list, buttons, and starting player
    global board, buttons, current_player, turn_label
    board = [" "] * 9
    current_player = "X"
    turn_label.config(text="Turn: X")

    # Clear button text and re-enable them
    for btn in buttons:
        btn.config(text=" ", state="normal", bg="#e6e6fa")


def check_winner_ui():
    # All winning combinations (rows, columns, diagonals)
    wins = [
        (0,1,2), (3,4,5), (6,7,8),
        (0,3,6), (1,4,7), (2,5,8),
        (0,4,8), (2,4,6)
    ]

    # Check if any winning pattern is filled by same player
    for a, b, c in wins:
        if board[a] == board[b] == board[c] != " ":
            return True
    return False


def check_draw_ui():
    # Draw happens when no empty spaces remain
    return " " not in board


def on_click(index):
    # Handle button click for each cell
    global current_player

    # Only allow move if cell is empty
    if board[index] == " ":
        board[index] = current_player

        # Update button appearance based on player
        buttons[index].config(
            text=current_player,
            bg="#d1ffd6" if current_player == "X" else "#ffd6d6",
            fg="#333333"
        )

        # Check for win
        if check_winner_ui():
            messagebox.showinfo("Winner", f"Player {current_player} wins!")
            # Disable all buttons after win
            for btn in buttons:
                btn.config(state="disabled")
            return

        # Check for draw
        if check_draw_ui():
            messagebox.showinfo("Draw", "It's a draw!")
            return

        # Switch player turn
        current_player = "O" if current_player == "X" else "X"
        turn_label.config(text=f"Turn: {current_player}")


# --- UI Setup ---

window = tk.Tk()
window.title("Tic Tac Toe")
window.config(bg="#dcd6f7")  # Background color

# Title label at top
title_label = tk.Label(window, text="TIC TAC TOE", font=("Arial", 28, "bold"),
                       bg="#dcd6f7", fg="#4b3f72")
title_label.pack(pady=10)

# Label showing whose turn it is
turn_label = tk.Label(window, text="Turn: X", font=("Arial", 16),
                      bg="#dcd6f7", fg="#4b3f72")
turn_label.pack(pady=5)

# Frame to hold the 3×3 grid
frame = tk.Frame(window, bg="#4b3f72", padx=4, pady=4)
frame.pack()

# Initialize board and buttons
board = [" "] * 9
current_player = "X"
buttons = []

# Create 9 buttons for the grid
for i in range(9):
    btn = tk.Button(
        frame,
        text=" ",
        font=("Arial", 26, "bold"),
        width=4,
        height=2,
        bg="#e6e6fa",
        activebackground="#cce7ff",
        relief="flat",
        command=lambda i=i: on_click(i)  # Pass index to click handler
    )
    btn.grid(row=i//3, column=i%3, padx=3, pady=3)
    buttons.append(btn)

# Reset button to restart the game
reset_btn = tk.Button(window, text="Reset Game", font=("Arial", 16),
                      bg="#f0e6ff", fg="#4b3f72", command=reset_board)
reset_btn.pack(pady=10)

window.mainloop()
```
