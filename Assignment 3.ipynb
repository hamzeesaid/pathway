3. Build a simple calculator which can: add, subtract, multiply, divide.
Hint: solve by writing a function that takes as argument two numbers and the operation and
returns the desired output.
Run test for function(2,5,’d’)
Output: 0.4

```python
def calculator(a, b, operation):
    if operation == 'a': #add
        return a + b
    elif operation == 's': #subtract
        return a - b
    elif operation == 'm': #multiply
        return a * b
    elif operation == 'd': # divide
        return a / b

calculator(2, 5, 'd') # 2 numbers and the the operation 
```

4. Define a class named Rectangle which can be constructed by a length and width.
The Rectangle class has a method which can compute the area.
Run test for r = Rectangle(5,10)
r.area()
 Output: 50


```python
class Rectangle:
    def __init__(self, length, width):
        
        self.length = length   ## Assign length to the object
        self.width = width     ## Assign Width to the object

    def area(self):
        return self.length * self.width ##Area formula: length × width

r = Rectangle(5, 10)  ## Create an object 'r' of the Rectangle class with length = 5 and width = 10

r.area() ##Call the area method;
```
